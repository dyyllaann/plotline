```python
# Setup
import numpy as np
from skimage.io import imread
from skimage.transform import resize
import matplotlib.pyplot as plt
import os

IMG_SIZE = (224, 224)
```

```python
# Load dataset
img = imread(os.path.join(FOLDERNAME, 'test.jpg'), as_gray=True)
plt.imshow(img)
```

```python
# Preprocess data
img_resized = resize(img, IMG_SIZE)
plt.imshow(img_resized)
```

```python
# HOG feature extraction
# Intuition is that mold will have edges in many directions, while root hairs will have edges in mostly one direction
from hog import hog_descriptor

hog_features = hog_descriptor(img_resized)
print("HOG feature vector length:", len(hog_features))
```

```python
# Canny feature extraction
# Intuition is edge detection and density
from edge import canny

edges = canny(img_resized, kernel_size=5, sigma=1.4, high=0.03, low=0.02)
print (edges.shape)

edge_density = np.sum(edges) / edges.size

plt.imshow(edges)
plt.axis('off')
```

```python
# Combine into feature vector
feature_vector = np.concatenate([hog_features, [edge_density]])
print("Combined feature vector length:", len(feature_vector))
```
